```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-regioni/dpc-covid19-ita-regioni.csv")
dfMolise = df.loc[df['denominazione_regione'] == "Molise"]
dfSardegna = df.loc[df['denominazione_regione'] == "Sardegna"]
```

```python
dfSardegna = dfSardegna.drop(columns=['note', 'note_test', 'note_casi'])
```
